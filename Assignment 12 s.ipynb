```python
import pandas as pd
```

```python
df = pd.read_csv('retail supermarket - retail supermarket.csv')
```

```python
rec = df[(df['Sales'] > 500) & (df['Quantity'] > 5)]
rec
```

```python
rec_2 = df[(df['Discount'] > 0.5) | (df['Profit'] < 0)]
rec_2
```

```python
rec_3 = df.loc[df['Region'] == 'East', ['Ship Mode', 'Profit']]
rec_3
```

```python
rec_4 = df[(df['Segment'] == 'Corporate') & (df['Region'] == 'Central')]
rec_4
```

```python
rec_5 = df[(df['Category'] == 'Furniture') & (df['Sub-Category'] != 'Chairs')]
rec_5
```

```python
rec_6 = df[(df['Discount'] > 0.3) & (df['Profit'] > 0)]
rec_6
```

```python
rec_7 =  df[(df['Profit'] > 500) & (df['Quantity'] > 5) & (df['Category'] == 'Technology')]
rec_7
```

```python
rec_8 = df.iloc[0:5, [df.columns.get_loc('Ship Mode'), df.columns.get_loc('Segment'), df.columns.get_loc('Sales')]]
rec_8
```

```python
rec_9 = df[(df['Segment'] == 'Home Office') & (df['Sales'] > 1000) & (df['Quantity'] > 3)]
rec_9
```

```python
df['Profit Margin'] = df['Profit'] / df['Sales']
rec_10 = df[df['Profit Margin'] > 0.3]
rec_10
```
