```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
df = pd.read_csv("Salaries - Salaries.csv")
df
```

```python
df = df[~df['EmployeeName'].str.contains('Not provided', case=False, na=False)]
df = df[~df['JobTitle'].str.contains('Not provided', case=False, na=False)]
df
```

```python
cols = ['BasePay','OvertimePay','OtherPay','Benefits']
df[cols] = df[cols].replace('Not Provided', 0)
df
```

```python
df[cols] = df[cols].astype(float)
df.info()
```

```python
df[df['BasePay'] > 200000][['EmployeeName','BasePay']]
```

```python
df[df.JobTitle.str.contains('MANAGER')]
```

```python
df[['EmployeeName','JobTitle','TotalPay']]
```

```python
(df['OvertimePay'] == 0).sum()
```

```python
df[df['TotalPay'] > 300000].groupby('JobTitle')['TotalPay'].sum()
```

```python
df.groupby('JobTitle')['TotalPayBenefits'].mean().nlargest(5)
```

```python
df['Benefits'] = df['Benefits'].fillna(0)
df.head()
```
