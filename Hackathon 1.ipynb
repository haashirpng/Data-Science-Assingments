```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('ecomerece data.csv')
df.head()
```

```python
df.info()
```

```python
df = df.drop(['Latitude', 'Longtitude'], axis=1)
```

```python
df = df.rename(columns ={'Order Priority Code':'Order Priority'})
df
```

```python
products = df.groupby('product')['total'].sum().sort_values(ascending=False).head(5)
print(products)
```

```python
sns.barplot(x=products.values,y=products.index,palette="viridis")
plt.title(" Top 5 Selling Products ")
```

## OBSERVATION 
The bar chart shows the top 5 selling products, in which headphones are  the most sold product, followed closely by Watch and Mobile.

```python
payment = df.payment_method.value_counts()
payment
```

```python
plt.figure(figsize=(7,7))
plt.pie(payment,labels=payment.index, autopct='%1.1f%%', startangle=90)
plt.title(" Payment Method Distribution ")
```

## OBSERVATION 
The pie chart  shows the Payment Method Distribution used by the customer, with JazzCash being the most frequently used payment Method followed closely by Credit Card and Easypaisa.

```python
monthwise = df.groupby('Month')['total'].sum().sort_values(ascending=False)
monthwise
```

```python
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
monthwise = monthwise[month_order]
plt.figure(figsize=(15, 6))
plt.plot(monthwise.index, monthwise.values,marker='D')
plt.title(" Month wise sales ")
```

## OBSERVATION 
The above line  Chart shows the Monthwise Sales pattern, and according to the chart, the highest sales were made in  March, and Sales generally remain high from January through May and then drop significantly during the summer months (June to September).

```python
city_mapping = {
    1.0: "Karachi",
    2.0: "Lahore",
    3.0: "Islamabad",
    4.0: "Faisalabad",
    5.0: "Quetta",
    6.0: "Peshawar"
}
df['City'] = df['City'].map(city_mapping)
```

```python
citywise = df.groupby('City')['total'].sum().sort_values(ascending=False)
citywise
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x=citywise.values, y=citywise.index,palette="viridis")
plt.title(" City wise sales ")
plt.ylabel("City")
```

## OBSERVATION 
The above bar chart shows the City-wise sales of the products, and according to the chart, the highest sales were made in the city of Lahore. Quetta & Islamabad also show strong sales, whereas Karachi records the lowest total sales among the cities.

```python
category = df.groupby(['Month', 'category'])['total'].sum().sort_values(ascending=False)
category
```

```python
monthly = df.pivot_table(index='Month', columns='category', values='total', aggfunc='sum')
monthly.plot(kind='bar', figsize=(10,6))
plt.title("Monthly Sales by Category")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.legend(title="Category")
```

## OBSERVATION 
The above chart shows Monthly Sales by Category. Accessories and Electronics drive the highest sales. The chart shows that Accessories is the top-performing category in March and January, which aligns with the overall peak sales period. Fashion shows a very high peak in February. Electronics generally maintains a consistent presence, peaking in March and April. 

## CONCLUSION 
The Sales Analysis shows that Headphones and Watches are top-selling products,which clearly shows the high demand in the Electronics and Accessories categories. Sales exhibit a noticeable seasonal trend, peaking sharply in March and maintaining high levels early in the year. Lahore leads in total sales, while JazzCash is the most preferred payment method overall. This also shows that most of the customers prefer online payment instead of COD.
The Analysis can help improve business decisions by Inventory & Marketing strategy,Seasonal Strategy and Payment Optimization.
